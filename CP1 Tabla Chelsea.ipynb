```python
#Nos aseguramos de que las celdas con texto no tengan espacios al final ni al principio
chelsea = chelsea.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#Las columnas con valores numericos las transformamos a formato int, en caso se necesite realizar algun cálculo
chelsea['P_Jugados'] = chelsea['P_Jugados'].astype(int)
chelsea['P_Ganados'] = chelsea['P_Ganados'].astype(int)
chelsea['P_Empatados'] = chelsea['P_Empatados'].astype(int)
chelsea['P_Perdidos'] = chelsea['P_Perdidos'].astype(int)
chelsea['Puntos'] = chelsea['Puntos'].astype(int)
chelsea['Posicion'] = chelsea['Posicion'].astype(int)
```

```python
#Para añadir información extra creamos una nueva columna que indique el porcentage de victorias 
#primero creamos una función personalizada para sacar el porcentaje
def porcentaje(x, y):
    return round((x / y) *100, 2) #redondemos a dos decimales
#Ahora creamos nueva columna aplicando la función personalizada
chelsea['pct_de_Victorias'] = chelsea.apply(lambda row: porcentaje(row['P_Ganados'], row['P_Jugados']), axis=1)
#Mostramos resultados
chelsea
```
